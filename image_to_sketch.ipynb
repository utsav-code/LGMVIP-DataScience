```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # To display images in Colab
```

```python
#Step 2 :  Read the image in RGB format

def read_image(file_path):
    image = cv2.imread(file_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image
```

```python
#step 3: Convert the image to grayscale

def convert_to_grayscale(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    return gray_image
```

```python
#Step 4: Invert the grayscale image (Negative Image)

def invert_image(image):
    inverted_image = 255 - image
    return inverted_image
```

```python
#Step 5: Create a pencil sketch by mixing the grayscale image with the inverted blurry image
def pencil_sketch(image):
    inverted_image = invert_image(image)
    blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), sigmaX=0, sigmaY=0)
    pencil_sketch_image = cv2.divide(image, 255 - blurred_image, scale=256)
    return pencil_sketch_image
```

```python
#Step 6: Display the original image and the pencil sketch side by side

def display_images(original_image, sketch_image):
    cv2_imshow(original_image)  # Display the original image
    cv2_imshow(sketch_image)    # Display the sketch image
```

```python
#Step 7: Finally running the code

if __name__ == "__main__":
    input_image_path = "/content/discord.png"  # Replace with the actual path to your image file
    original_image = read_image(input_image_path)
    grayscale_image = convert_to_grayscale(original_image)
    sketch_image = pencil_sketch(grayscale_image)
    display_images(original_image, sketch_image)
```
